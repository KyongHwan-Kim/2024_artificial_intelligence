```python
# 2.1    
import numpy
import matplotlib.pyplot as plt
import pandas
import random

# Load the data
data = pandas.read_csv('Grid_Data.csv')
data.columns = ['x', 'y']

# Set the initial value
x = data['x']
y = data['y']
w = 0
b = 0
f = [(w*num + b) for num in x]
mse = numpy.mean((f - y)**2)

# epoch , w,b    
def fitting_function(variable):
    init = random.uniform(-10, 10)
    return init

# MSE     
def find_optimal_params(history):
    for e in history:
        if e[3] == min(history, key = lambda t: t[3])[3]:
            return e

def draw_graph(w, b, mse):
    f = [(w*num + b) for num in x]
    plt.title('My Graph')
    # draw the fitting line
    plt.plot(x, f)
    # draw the original data
    plt.plot(x, y, 'o')
    # show annotation
    plt.text(1, 11 ,'MSE : ' + str(round(mse, 2)), color='blue')
    plt.text(1, 10.5 ,'w : ' + str(round(w, 2)), color='green')
    plt.text(1, 10 ,'b : ' + str(round(b, 2)), color='gold')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True, alpha=0.5)
    # show the graph
    plt.show()
    

def draw_mse_graph(history, optimal_params):
    plt.title('MSE Graph')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.plot([i[0] for i in history], [i[3] for i in history])
    plt.plot(optimal_params[0], optimal_params[3], 'ro')
    plt.grid(True, alpha=0.5)
    plt.annotate('Optimal \n('+ str(optimal_params[0]) + ', ' + 
                 str(round(optimal_params[3], 2)) +')', 
                 xy=(optimal_params[0] - 5, optimal_params[3] - 5),
                 xytext=(optimal_params[0] - 10, optimal_params[3] - 10))
    plt.show()

def main():
    history = []
    for idx in range(0, 100):
        w = fitting_function('w')
        b = fitting_function('b')
        f = [(w*num + b) for num in x]
        mse = numpy.mean((f - y)**2)
        history.append([idx, w, b, mse])
    optimal_params = find_optimal_params(history)
    # draw the graph
    draw_graph(optimal_params[1], optimal_params[2], optimal_params[3])
    draw_mse_graph(history, optimal_params)
    
if __name__ == "__main__":
    main()
```

```python
# 2.4     

import numpy
import matplotlib.pyplot as plt
import pandas
import random

# Load the data
data = pandas.read_csv('Grid_Data.csv')
data.columns = ['x', 'y']

x = data['x']
y = data['y']
w = 0
b = 0
f = [(w*num + b) for num in x]
mse = numpy.mean((f - y)**2)

# epoch , w,b    
def fitting_function(history):
    # history   :     (w : 0~10, b : -5~5)
    if len(history) == 0:
        w = random.uniform(0, 10)
        b = random.uniform(-5, 5)
        return w, b
    #  MSE  
    else:
        sum_mse = sum([e[3] for e in history])
        sum_w = sum([e[1] * (1 - ( e[3] / sum_mse )) for e in history])
        sum_b = sum([e[2] * (1 - ( e[3] / sum_mse )) for e in history])
        w = sum_w / len(history) + random.uniform(-1, 1)
        b = sum_b / len(history) + random.uniform(-1, 1)
        return w, b
    
# MSE     
def find_optimal_params(history):
    for e in history:
        if e[3] == min(history, key = lambda t: t[3])[3]:
            return e

def draw_graph(w, b, mse):
    f = [(w*num + b) for num in x]
    plt.title('My Graph')
    # draw the fitting line
    plt.plot(x, f)
    # draw the original data
    plt.plot(x, y, 'o')
    # show annotation
    plt.text(1, 11 ,'MSE : ' + str(round(mse, 2)), color='blue')
    plt.text(1, 10.5 ,'w : ' + str(round(w, 2)), color='green')
    plt.text(1, 10 ,'b : ' + str(round(b, 2)), color='gold')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True, alpha=0.5)
    # show the graph
    plt.show()
    
def draw_mse_graph(history, optimal_params):
    plt.title('MSE Graph')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.plot([i[0] for i in history], [i[3] for i in history])
    plt.plot(optimal_params[0], optimal_params[3], 'ro')
    plt.grid(True, alpha=0.5)
    plt.annotate('Optimal \n('+ str(optimal_params[0]) + ', ' + 
                 str(round(optimal_params[3], 2)) +')', 
                 xy=(optimal_params[0] - 5, optimal_params[3] - 5),
                 xytext=(optimal_params[0] - 10, optimal_params[3] - 10))
    plt.show()

def main():
    history = []
    for idx in range(0, 100):
        w, b = fitting_function(history)
        f = [(w*num + b) for num in x]
        mse = numpy.mean((f - y)**2)
        history.append([idx, w, b, mse])
    optimal_params = find_optimal_params(history)
    # draw the graph
    draw_graph(optimal_params[1], optimal_params[2], optimal_params[3])
    draw_mse_graph(history, optimal_params)
    
if __name__ == "__main__":
    main()
```
